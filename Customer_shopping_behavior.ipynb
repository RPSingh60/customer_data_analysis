```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
df['review_rating'] = df['review_rating'].fillna(df.groupby('category')['review_rating'].transform('median'))
```

```python
df = df.rename(columns = {'purchase_amount_(usd)' : 'purchase_amount'})
```

```python
labels = ['yong_adult','mid_aged','senior','middle-aged']
df['age_group'] = pd.qcut(df['age'], q = 4, labels = labels)
```

```python
df
```

```python
df[['age','age_group']].head(10)
```

```python
frequency_mapping = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Bi-Weekly' : 14,
    'Quarterly' : 90,
    'Annually' : 365,
    'Monthly' : 30,
    'Every 3 Months' : 90
}
df['frequency_of_purchases_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
